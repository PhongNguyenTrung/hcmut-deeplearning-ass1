```python
import sys
sys.path.append('../src')
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import fetch_20newsgroups
from collections import Counter
print('Libraries loaded ✅')
```

## 1. Load Dataset

```python
train_raw = fetch_20newsgroups(subset='train', remove=('headers','footers','quotes'))
test_raw  = fetch_20newsgroups(subset='test',  remove=('headers','footers','quotes'))
print(f'Train: {len(train_raw.data)} | Test: {len(test_raw.data)}')
print(f'Categories: {train_raw.target_names}')
```

## 2. Class Distribution

```python
counts = Counter(train_raw.target)
class_names = train_raw.target_names
plt.figure(figsize=(14, 5))
plt.bar([class_names[i] for i in sorted(counts)], [counts[i] for i in sorted(counts)], color='coral')
plt.xticks(rotation=45, ha='right')
plt.title('20 Newsgroups – Class Distribution')
plt.tight_layout()
plt.savefig('../results/newsgroups_class_dist.png', dpi=150)
plt.show()
```

## 3. Text Length Distribution

```python
lengths = [len(text.split()) for text in train_raw.data]
plt.figure(figsize=(10, 4))
plt.hist(lengths, bins=50, color='teal', edgecolor='white')
plt.axvline(np.median(lengths), color='red', linestyle='--', label=f'Median: {np.median(lengths):.0f}')
plt.title('Distribution of Document Lengths (words)')
plt.xlabel('Words')
plt.ylabel('Count')
plt.legend()
plt.tight_layout()
plt.savefig('../results/newsgroups_length_dist.png', dpi=150)
plt.show()
print(f'Mean: {np.mean(lengths):.1f} | Median: {np.median(lengths):.1f} | Max: {max(lengths)}')
```

## 4. Sample Documents

```python
for i in range(3):
    print(f'--- Sample {i+1} | Label: {class_names[train_raw.target[i]]} ---')
    print(train_raw.data[i][:400])
    print()
```

## 5. Tokenization Statistics (DistilBERT)

```python
from transformers import AutoTokenizer
tokenizer = AutoTokenizer.from_pretrained('distilbert-base-uncased')
token_lengths = [len(tokenizer.encode(t, truncation=False)) for t in train_raw.data[:2000]]
print(f'Token lengths – Mean: {np.mean(token_lengths):.1f} | 95th pct: {np.percentile(token_lengths, 95):.1f}')
for threshold in [128, 256, 512]:
    pct = sum(l <= threshold for l in token_lengths) / len(token_lengths) * 100
    print(f'  <= {threshold} tokens: {pct:.1f}%')
```
